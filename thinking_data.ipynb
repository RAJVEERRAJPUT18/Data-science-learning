```python
import json
```

```python
def load_data(filename):
  data=[]
  with open(filename ,"r", encoding="utf-8") as f:
      data=json.load(f)
  return data
```

```python
data=load_data("Data.json")
print(data)
```

```python
def clean_data(data):
    text_to_num = {"one": 1, "two": 2, "three": 3, "four": 4, "five": 5}
    cleaned_data = []
    unique_user = set()

    for user in data:
        # Clean rating
        raw_rating = str(user["Rating"]).strip().lower()
        if raw_rating in text_to_num:
            raw_rating = text_to_num[raw_rating]
        user["Rating"] = raw_rating

        # Handle missing age
        raw_age = user.get("Age")
        if raw_age is None:          
            user["Age"] = None       

        # Deduplication
        if user["name"].strip() in unique_user: 
            continue
        unique_user.add(user["name"].strip())
        cleaned_data.append(user)

    return cleaned_data             
        
        
```

```python
data=clean_data(data)


```

```python
# meaningful insight
# 1. what is avegrage rating 
# 2. what % user of poor rtaing 
def get_insights(data):
# avg of rating
   total_rating=0
   for user in data:
       total_rating += float(user["Rating"])
   print(f"avg of rating ={total_rating/len(data)}")

#  caculate  the persentage of poor Rating
   poor_rating=0
   for user in data:
      if(float(user["Rating"]) >3):
             poor_rating+=1

   print(f"persentage of poor_rating ={poor_rating/len(data)*100}")
     
         
     
```

```python
get_insights(data)
```

```python
# Assignment 3
# Build a product recommendation feature (Rule based, not ML)
# 1.If user rating >= 4, recommend same brand products.
# 2.If user rating < 4, recommend different brand products.
def get_recomndation(data):
    recomndation=[]
    for user in data:
        curnt_recomn={}
        curnt_recomn["name"]=user["name"]
        if(float(user["Rating"])>=4):
            curnt_recomn["brand"]="Apple"
        else:
            curnt_recomn["brand"]="Samsung"
        recomndation.append(curnt_recomn)
    return recomndation
```
